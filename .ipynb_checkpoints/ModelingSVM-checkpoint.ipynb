```python
import pandas as pd 
import numpy as np 
import boto3
from boto3.session import Session 
pd.set_option('display.max_columns', 500)
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
import configparser
Config = configparser.ConfigParser()
Config.read_file(open('/Users/milesklingenberg/Documents/Personal/AWS/AWS_Keys'))
```

```python
KEY = Config.get('AWS', 'AWS_ACCESS_KEY_ID')
SECRET = Config.get('AWS', 'Secret')
```

```python
s3client = boto3.client('s3', 
                       aws_access_key_id=KEY,
                       aws_secret_access_key=SECRET)
file = s3client.get_object(Bucket = 'pubg-competition', Key = 'train_agg.csv')
```

```python
total_train = pd.read_csv(file['Body'])
total_train = pd.DataFrame(total_train)
```

```python
total_train.head(2)
```

```python
X = total_train[[ 'assists', 'boosts', 'damageDealt', 'DBNOs', 'headshotKills', 
               'heals', 'kills', 'killStreaks', 'longestKill', 'matchDuration', 'maxPlace', 
               'numGroups', 'rankPoints', 'revives', 'rideDistance', 'roadKills', 'swimDistance', 
               'teamKills', 'vehicleDestroys', 'walkDistance', 'weaponsAcquired', 'winPoints', 
               'max_kills', 'min_kills', 'max_walk', 'min_walk', 'max_dbnos', 'min_dbnos', 'ratio_kill', 
              'ratio_dbnos', 'ratio_walk', 'damage_kill_ratio', 'total_distance']]

y = total_train['winPlacePerc']
```

```python
from sklearn import model_selection

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, train_size=0.65,test_size=0.35, random_state=101)

X_train.head(2)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

y_pred = pd.DataFrame(y_pred)
rms = sqrt(mean_squared_error(y_test, y_pred))
print(rms)
```
